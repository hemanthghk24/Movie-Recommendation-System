```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
counter=CountVectorizer(max_features=10000,stop_words='english')
```

```python
counter
```

```python
vector=counter.fit_transform(new_list['genre'].values.astype('U')).toarray()
```

```python
import ast
movie['genre'] = movie['genre'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
import ast

def safe_parse_genre(x):
    if isinstance(x, list):
        return x
    if isinstance(x, str):
        try:
            parsed = ast.literal_eval(x)
            return parsed if isinstance(parsed, list) else [parsed]
        except (ValueError, SyntaxError):
            return [g.strip() for g in x.split(',')]
    return []

movie['genre'] = movie['genre'].apply(safe_parse_genre)
```

```python
import ast

def safe_parse_genre(x):
    if isinstance(x, list):
        return x
    if isinstance(x, str):
        try:
            parsed = ast.literal_eval(x)
            return parsed if isinstance(parsed, list) else [parsed]
        except (ValueError, SyntaxError):
            return [g.strip() for g in x.split(',')]
    return []

# Safely create a modifiable copy of the DataFrame before applying changes
movie = movie.copy()
movie['genres'] = movie['genre'].apply(safe_parse_genre)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_list[new_list['title']=="The Godfather"].index[0]
```

```python
distance= sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_list.iloc[i[0]].title)
```

```python
def recommand(movies):
    index=new_list[new_list['title']==movies].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_list.iloc[i[0]].title)
```

```python
recommand("Iron Man")
```
